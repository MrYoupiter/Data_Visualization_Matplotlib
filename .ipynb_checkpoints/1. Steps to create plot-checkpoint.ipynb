```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
%matplotlib inline
```

```python
#generate random numbers (total 10)
randomNumber = np.random.rand(10)
```

```python
print randomNumber
```

```python
style.use('ggplot')
plt.plot(randomNumber, 'g', label='line one', linewidth=2) #g is green
plt.xlabel('Range')
plt.ylabel('Numbers')
plt.title('First Plot')
plt.legend() #to plot the graph 
plt.show() #to display
```
