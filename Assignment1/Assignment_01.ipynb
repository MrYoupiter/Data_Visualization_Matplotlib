#### 2: Write a user-defined function for origin  


```python
#use apply function
#write a user defined function for origin
#1-USA, 2-Europe, 3-Asia
def origin(num):
    if num==1:
        return 'USA'
    elif num==2:
        return 'EUROPE'
    else:
        return 'Asia'
#use apply function
df_auto_dataset['origin'] = df_auto_dataset['origin'].apply(origin)
```

```python
#view first 30 data points 
df_auto_dataset.head(30)
```

#### 3: Draw the pair plot using sns for mpg, weight,orign and with hue origin, set the size to 4


```python
#hue is variable in data set to map plot aspects in different colors
sns.pairplot(df_auto_dataset[['mpg', 'weight', 'origin']], hue='origin', size=4)
```
